```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("weapons.csv")
df

df["Weapon Type"].value_counts().head(10).plot(kind="pie", autopct="%1.1f%%", figsize=(6,6))
plt.title("Weapon Type")
plt.ylabel("")
plt.show()

df["Bullet Type"].value_counts().head(10).plot(kind="pie", autopct="%1.1f%%", figsize=(6,6))
plt.title("Bullet Type")
plt.ylabel("")
plt.show()

plt.scatter(df["Range"], df["Damage"])
plt.title("Damage vs Range")
plt.xlabel("Range")
plt.ylabel("Damage")
plt.show()


sns.boxplot(y=df["Damage"])
plt.title("Damage Boxplot")
plt.show()


sns.countplot(x="Fire Mode", data=df)
plt.xticks(rotation=90)
plt.title("Fire Mode Count")
plt.show()


sns.histplot(df["Damage Per Second"])
plt.title("Damage Per Second Distribution")
plt.xlabel("Damage Per Second")
plt.show()

sns.histplot(df["Bullet Speed"])
plt.title("guns")
plt.xlabel("speed")
plt.show()

sns.countplot(x="Weapon Type", data=df)
plt.xticks(rotation=90)
plt.title("Weapon Type Count")
plt.show()

top_salary = df.sort_values(by='Damage', ascending=True).head(10)

plt.figure(figsize=(8,5))
plt.barh(top_salary['Weapon Name'],
         top_salary['Damage'])

plt.title("Top 10 Highest damage smg")
plt.xlabel("Damage")
plt.ylabel("Weapon Name")
plt.show()
```
